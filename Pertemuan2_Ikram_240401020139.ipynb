```python
import pandas as pd
import numpy as np

# 1 Muat Dataset Titanic
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head()
```

```python
# 2 Eksplorasi Awal
print(df.shape, df.columns.tolist())
print(df.isnull().sum())
```

```python
# 3 Filtering — Penumpang Wanita Kelas 1
wk1 = df[(df['Sex']=='female') &
(df['Pclass']==1)]

print(wk1['Survived'].mean(), wk1['Age'].mean())
```

```python
# 4 Groupby — Tingkat Keselamatan per Kelas
hasil = df.groupby('Pclass').agg(
total=('PassengerId','count'),
selamat=('Survived','mean'),
).round(3)
print(hasil)
```

### 1. Berapa jumlah baris dan kolom dataset Titanic?

```python
print(f"Dataset Titanic memiliki {df.shape[0]} baris dan {df.shape[1]} kolom.")
```

### 2. Kolom mana yang memiliki nilai kosong (missing)?

```python
print("Kolom dengan nilai kosong:")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

### 3. Berapa persen penumpang yang selamat (Survived=1)?

```python
persentase_selamat = df['Survived'].mean() * 100
print(f"Persentase penumpang yang selamat adalah: {persentase_selamat:.2f}%")
```

### 4. Rata-rata usia penumpang wanita kelas 1 yang selamat?

```python
rata_rata_usia_wanita_kelas1_selamat = df[(df['Sex'] == 'female') & (df['Pclass'] == 1) & (df['Survived'] == 1)]['Age'].mean()
print(f"Rata-rata usia penumpang wanita kelas 1 yang selamat adalah: {rata_rata_usia_wanita_kelas1_selamat:.2f} tahun")
```

### 5. Kelas penumpang mana yang tingkat keselamatan tertinggi?

```python
hasil_selamat_per_kelas = df.groupby('Pclass')['Survived'].mean().reset_index()
kelas_selamat_tertinggi = hasil_selamat_per_kelas.loc[hasil_selamat_per_kelas['Survived'].idxmax()]
print(f"Kelas penumpang dengan tingkat keselamatan tertinggi adalah Pclass {int(kelas_selamat_tertinggi['Pclass'])} dengan tingkat selamat {kelas_selamat_tertinggi['Survived']:.3f}")
```

### 6. Tampilkan hasil `groupby('Pclass')['Survived'].mean()`

```python
print(df.groupby('Pclass')['Survived'].mean().round(3))
```
